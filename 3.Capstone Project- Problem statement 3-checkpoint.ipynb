```python
#Data Preprocessing and Transformation
df = pd.DataFrame(df,columns = ['Score','Text'])
df.rename(columns = {'Score':'Rating','Text':'Review'},inplace = True)
```

```python
def apply_sentiment(Rating):
    if(Rating <=2 ):
        return 0
    else:
        return 1
```

```python
df['Sentiment'] = df['Rating'].apply(apply_sentiment)
df.drop(['Rating'],axis = 1, inplace = True)
```

```python
sentiment = df['Sentiment'].value_counts()
sentiment
```

```python
plt.figure(figsize = ( 10,8))
sns.countplot(data = df, x = 'Sentiment')
plt.show()
```

```python
def clean_text(Review):
   
    Review = str(Review).lower() # convert to lowercase
    Review = re.sub('\[.*?\]', '', Review) 
    Review = re.sub('https?://\S+|www\.\S+', '', Review) # Remove URls
    Review = re.sub('<.*?>+', '', Review)
    Review = re.sub(r'[^a-z0-9\s]', '', Review) # Remove punctuation
    Review = re.sub('\n', '', Review)
    Review = re.sub('\w*\d\w*', '', Review)
    return Review
```

```python
df['Review'] = df['Review'].apply(clean_text)
```

```python
df.head()
```

```python
df.Review=df.Review.apply(nltk.word_tokenize)
```

```python
df.Review
```

```python
from nltk.stem import PorterStemmer
```

```python
stem_obj=PorterStemmer()
```

```python
df.Review=df.Review.apply(lambda row: [stem_obj.stem(each_word) for each_word in row])
```

```python
df.Review[0]
```

```python
df.Review=df.Review.apply(lambda row:" ".join(row))
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
obj=CountVectorizer()
```

```python
df.reviews_text=obj.fit_transform(df.Review)
```

```python
df.reviews_text
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.reviews_text, df.Sentiment,random_state = 42,
                                                   test_size = 0.20)
X_train.shape,X_test.shape,y_train.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
clf = Pipeline([
    ('tfidf', TfidfTransformer()),
    ('classifier', MultinomialNB()),
    ])
```

```python
fit_model = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print('Training accuracy:', fit_model.score(X_train,y_train))
print('Test accuracy:', fit_model.score(X_test,y_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))

print(confusion_matrix(y_test,y_pred))
```
